```python
import pandas as pd

# Cargar datos desde el archivo CSV
file_path = './graph/matrix_algorithms_data.csv'  # Cambia al nombre de tu archivo
data = pd.read_csv(file_path)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
matrix_data = pd.read_csv('./graph/matrix_algorithms_data.csv')
list_data = pd.read_csv('./graph/list_algorithms_data.csv')
list_data["Structure"] = "Adjacency List"
matrix_data["Structure"] = "Adjacency Matrix"
combined_data = pd.concat([list_data, matrix_data], ignore_index=True)

# Sumar el tiempo total por algoritmo y estructura
total_time = combined_data.groupby(['Algorithm', 'Structure'], as_index=False)['Time(ms)'].sum()

# Graficar la comparacin de los tiempos totales
plt.figure(figsize=(12, 6))
sns.barplot(
    data=total_time,
    x="Algorithm",
    y="Time(ms)",
    hue="Structure"
)
plt.title("Comparacin de tiempos totales por algoritmo y estructura de adyacencia")
plt.xlabel("Algoritmo")
plt.ylabel("Tiempo total de ejecucin (ms)")
plt.xticks(rotation=45, ha="right")
plt.legend(title="Estructura")
plt.grid(True)
plt.yscale("log")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

algos = data['Algorithm'].unique()

for algo in algos:
    # Filtrar por un algoritmo, por ejemplo "Dijkstra"
    dijkstra_data = data[data['Algorithm'] == algo]

    # Crear el grfico
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=dijkstra_data, x='Density', y='Time(ms)', hue='Vertices', marker='o')

    plt.title("Tiempo de " + algo + " vs Densidad del Grafo", fontsize=15)
    plt.xlabel('Densidad del Grafo', fontsize=12)
    plt.ylabel('Tiempo (ms)', fontsize=12)
    plt.legend(title='Vertices')
    plt.yscale('log')
    plt.grid(True)
    plt.show()
```

```python
densities = data['Density'].unique()

for density in densities:
    # Filtrar por densidad fija, por ejemplo 1
    density_1_data = data[data['Density'] == density]

    # Crear el grfico
    plt.figure(figsize=(12, 8))
    sns.barplot(data=density_1_data, x='Algorithm', y='Time(ms)', hue='Vertices')

    plt.title('Comparacin de Algoritmos a Densidad ' + str(density), fontsize=15)
    plt.xlabel('Algoritmo', fontsize=12)
    plt.ylabel('Tiempo (ms)', fontsize=12)
    plt.xticks(rotation=45)
    plt.legend(title='Vertices')
    plt.yscale('log')
    plt.show()

```

```python
algos = data['Algorithm'].unique()
for algo in algos:
    fw_data = data[data['Algorithm'] == algo]

    # Crear el grfico
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=fw_data, x='Vertices', y='Time(ms)', hue='Density', marker='o')

    plt.title('Tiempo de ' + algo + ' vs Nmero de Vrtices', fontsize=15)
    plt.xlabel('Nmero de Vrtices', fontsize=12)
    plt.ylabel('Tiempo (ms)', fontsize=12)
    plt.legend(title='Densidad')
    plt.grid(True)
    plt.show()

```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(data=data, x='Density', y='Time(ms)', hue='Algorithm', style='Vertices', markers=True)

plt.title('Comparacin de Algoritmos en Funcin de la Densidad', fontsize=14)
plt.xlabel('Densidad del Grafo', fontsize=12)
plt.ylabel('Tiempo (ms)', fontsize=12)
plt.legend(title='Algoritmo', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.yscale('log')
plt.grid(True)
plt.show()
```
